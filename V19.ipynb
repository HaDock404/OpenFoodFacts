## B. Traitement des valeurs manquantes
Le traitement des valeurs manquantes est une étape importante de notre étude car ces valeurs ne permettent pas d'observer la réprésentation des variables de chaque individus.

#### Lorsque le nom du produit n'est pas renseigné il est impossible de le retrouver de quelque manière que ce soit, il est donc préférable de le supprimer pour simplifier notre étude

```python
index_null_value = new_df[new_df['product_name'].isnull()].index
new_df.drop(index_null_value , inplace=True)
print((len(index_null_value)/len(new_df)*100, 2))
```

#### Lorsque aucune données ne sont renseignées pour un individu il est préférable de le supprimer pour faciliter notre travail.
#### On affiche également le pourcentage de suppression dans le Dataset

```python
df_index_drop = []
for i in new_df.itertuples():
    if pd.isna(i.energy_100g)\
    & pd.isna(i.proteins_100g)\
    & pd.isna(i.sugars_100g)\
    & pd.isna(i.fat_100g)\
    & pd.isna(i.salt_100g)\
    & pd.isna(i.sodium_100g)\
    & pd.isna(i.fiber_100g)\
    & pd.isna(i.saturated_fat_100g)\
    & pd.isna(i.carbohydrates_100g)\
    & pd.isna(i.nutrition_grade_fr):
        df_index_drop.append(i.Index)
        
len_list = len(df_index_drop)
print(round(len_list/len(new_df)*100, 2))
df = new_df.drop(df_index_drop).copy()
```

#### Comme nous pouvons le constater certaines variables n'ont pas pu être encore traitée, il faudra alors effectuer un nettoyage plus complexe pour réussir à déterminer les valeurs manquantes de ces variables ou les supprimer

```python
msno.matrix(df.sample(250))
```

#### Pour parvenir à retrouver les valeurs manquantes nous pourrons utiliser une fonction IterativeImputer, mais pour cela il faut qu'il y ai un corrélation entre les variables pour pouvoir applique ce type de fonction.

#### On commence par utiliser seulement les individus n'ayant pas de données manquantes dans leur variable pour détecter les corrélations qu'ils pourraient avoir entre eux

```python
df_index_drop = []
for i in df.itertuples():
    if pd.isna(i.energy_100g)\
    | pd.isna(i.proteins_100g)\
    | pd.isna(i.sugars_100g)\
    | pd.isna(i.fat_100g)\
    | pd.isna(i.salt_100g)\
    | pd.isna(i.sodium_100g)\
    | pd.isna(i.fiber_100g)\
    | pd.isna(i.saturated_fat_100g)\
    | pd.isna(i.carbohydrates_100g)\
    | pd.isna(i.nutrition_grade_fr):
        df_index_drop.append(i.Index)
        
df_without_Nan = df.drop(df_index_drop).copy()
```

#### Puis nous représentons graphiquement une heatmap de la corrélation de Pearson qui permet de détecter les corrélations entre différentes variables lorsqu'une case s'approche de la couleur rouge.

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df_without_Nan.corr()

plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap de la Corrélation de Pearson')
plt.show()
```

#### Nous pouvons alors constater une corrélation entre les variables 'salt_100g' et 'sodium_100g' également entre les variables 'fat_100g',  'satured_fat_100g' et 'energy_100g',  et enfin entre les variables 'carbohydrates_100g', 'sugars_100g' et 'energy_100g'

#### Nous utilisons alors la fonction IterativeImputer pour imputer aux valeurs manquantes une valeurs interdépendantes avec sa variable de corrélation 

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
df['fat_100g'] = imputer.fit_transform(df[['fat_100g', 'saturated_fat_100g',\
                                           'energy_100g']])[:, 0]
df['saturated_fat_100g'] = imputer.fit_transform(df[['saturated_fat_100g', 'fat_100g',\
                                                     'energy_100g']])[:, 0]
df['energy_100g'] = imputer.fit_transform(df[['energy_100g', 'fat_100g', 'saturated_fat_100g',\
                                              'carbohydrates_100g']])[:, 0]
df['carbohydrates_100g'] = imputer.fit_transform(df[['carbohydrates_100g', 'sugars_100g',\
                                                     'energy_100g']])[:, 0]
df['sugars_100g'] = imputer.fit_transform(df[['sugars_100g', 'carbohydrates_100g']])[:, 0]
```

#### Malheureusement cette imputation fait apparaitre de nouvelles valeurs aberrantes pour la variable fat_100g, energy_100g, carbohydrates_100g et sugars_100g que nous devons de nouveau supprimer.

```python
index_drop_rogue_value = df[df['fat_100g'] < 0].index
df.drop(index_drop_rogue_value, inplace=True)
index_drop_rogue_value = df[df['fat_100g'] > 100].index
df.drop(index_drop_rogue_value, inplace=True)
index_drop_rogue_value = df[df['carbohydrates_100g'] > 100].index
df.drop(index_drop_rogue_value, inplace=True)
index_drop_rogue_value = df[df['energy_100g'] > 3619.5].index
df.drop(index_drop_rogue_value, inplace=True)
index_drop_rogue_value = df[df['sugars_100g'] < 0].index
df.drop(index_drop_rogue_value, inplace=True)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['fat_100g'], df['saturated_fat_100g'])
plt.xlabel('fat_100g')
plt.ylabel('saturated_fat_100g')
plt.title('Nuage de points : fat_100g vs saturated_fat_100g')

plt.show()
```

#### Nous pouvons supprimer ces valeurs en traçant visuellement une droite qui restreint notre modèle à des valeurs cohérentes avec la corrélation des variables.

```python
a = 1
b = 10

df = df[df['saturated_fat_100g'] <= a * df['fat_100g'] + b]

plt.figure(figsize=(8, 6))
plt.scatter(df['fat_100g'], df['saturated_fat_100g'])
plt.xlabel('fat_100g')
plt.ylabel('saturated_fat_100g')
plt.title('Nuage de points : fat_100g vs saturated_fat_100g')

plt.show()
```

#### Pour imputer des valeurs correspondantes à cette variable nous allons faire appel à un algorithme de K-Nearest-Neighbor

```python
import pandas as pd
import numpy as np
from numpy import nan

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

#outils pour tester notre modèle
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

#### Nous utilisons seulement les individus qui ont toutes leurs valeurs pour entrainer l'algorithme KNN

```python
df_index_drop = []
for i in df.itertuples():
    if pd.isna(i.nutrition_grade_fr):
        df_index_drop.append(i.Index)
        
df_without_Nan = df.drop(df_index_drop).copy()
```

#### Nous allons prendre un échantillon de 50.000 individus pour accélerer l'entrainement de notre modèle

```python
df_without_Nan.shape
```

```python
df_sample = df_without_Nan.sample(50000, replace=False)
```

#### Nous prenons les variables sur lesquels entrainer notre modèle et la variable que l'on souhaite retrouver
#### Nous divisons notre dataframe en un échantillon d'entrainement et un échantillon de 20% pour tester notre modèle

```python
X = df_sample.iloc[:, 2:11]
y = df_sample.iloc[:, 1]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

#### Nous mettons à l'échelle nos données 

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

#### Nous observons pour quel nombre de plus proches voisins il y a un bon compromis entre biais et variance, en l'occurence ici nous prendrons 3 plus proches voisins

```python
errors = []
for k in range(2,15):
    knn = KNeighborsClassifier(k)
    errors.append(100*(1 - knn.fit(X_train, y_train).score(X_test, y_test)))
plt.plot(range(2,15), errors, 'o-')
plt.show()
```

#### Nous définissons notre modèle de classifieur 

```python
classifier = KNeighborsClassifier(n_neighbors=3, p=2, metric='euclidean')
```

#### Nous adaptons notre classifier sur notre échantillon

```python
classifier.fit(X_train, y_train)
```

#### Nous effectuons une prédictons sur la variable 'nutrion_grade_fr' avec le classifieur que nous avons paramétré juste au dessus

```python
y_pred = classifier.predict(X_test)
y_pred
```

#### Nous vérifions le pourcentage de correspondance entre nos données prédite et ce que nous avions initialement dans le dataset pour vérifier la pertinence de notre modèle

```python
print(accuracy_score(y_test, y_pred))
```

#### Nous allons désormais pouvoir appliquer notre classifieur sur nos données manquantes
#### Nous conservons les individus dont il ne manque pas de valeurs pour l'entrainement de notre modèle, nous mettons à l'échelle nos échantillons puis nous paramétrons notre classifieur et enfin nous l'adaptons à nos données d'entrainement.

```python
X = df_without_Nan.iloc[:, 2:11]
y = df_without_Nan.iloc[:, 1]
```

```python
X_train = X
y_train = y
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3, p=2, metric='euclidean')
```

```python
classifier.fit(X_train, y_train)
```

#### Nous utilisons notre modèle pour prédire nos données manquantes

```python
X_test = df.iloc[:, 2:11]
values_predicted = classifier.predict(X_test)
```

## C. Analyse multivariées

Désormais nous pouvons effectuer une analyse descriptive multivarié en commencant par l'analyse en composante principale (ACP).

#### Pour simplifier l'étude de notre Dataset, nous avons transformé l'index du Dataset en le nom de chaque individu

```python
df.set_index('product_name', inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

#### Nous supprimons la variable nutrition_grade_fr pour notre analyse en composante principale car cette variable correspond à un objet

```python
df_PCA = df.drop('nutrition_grade_fr', axis=1)
```

#### Nous standardisons nos données (centré-réduit)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_PCA)
```

#### Nous vérifions que la moyenne est de 0 et l'écart type de 1 avec nos données centrées réduites

```python
idx = ["mean", "std"]

pd.DataFrame(X_scaled).describe().round(2).loc[idx, :]
```

#### Nous allons travailler sur l'ensemble des variables de notre Dataframe

```python
n_components = 9
```

#### Nous instancions notre ACP avec le nombre de variable de notre dataset que nous souhaitons observer.

```python
pca = PCA(n_components=n_components)
```

#### Et enfin nous entrainons nos données sur les données standardisé (centrées-réduites)

```python
pca.fit(X_scaled)
```

#### Nous pouvons désormais nous intéresser à la variance captée par chaque nouvelles composantes et afficher son pourcentage par rapport à la variance totale dans le dataset

```python
pca.explained_variance_ratio_
```

```python
scree = (pca.explained_variance_ratio_*100).round(2)
scree
```

#### La fonction cumsum nous permet d'avoir la somme cumulé de la variance de chaque composante, cela nous sera utile pour l'affichage graphique de la variance des composantes

```python
scree_cum = scree.cumsum().round()
scree_cum
```

#### Et enfin nous pouvons afficher graphiquement la répartition de la variance des composantes principales
#### Avec en bleu la variance de chaque nouvelle composante et en rouge la variance cumulée.

```python
x_list = range(1, n_components+1)
list(x_list)

plt.bar(x_list, scree)
plt.plot(x_list, scree_cum,c="red",marker='o')
plt.xlabel("rang de l'axe d'inertie")
plt.ylabel("pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

#### Avec l'attribut components_ nous obtenons une matrice qui représente les vecteurs propres (ou composantes principales) trouvés par l'analyse en composantes principales. Ces vecteurs propres correspondent aux axes principaux de variation dans l'espace des données d'origine.
#### Enfin nous transposons la matrice pour effectuer la multiplication entre la matrice des données originales et la matrice des composantes principales afin que la dimension des matrices corresponde correctement pour nous permettre de représenter graphiquement les corrélations des différentes variables entre les composantes principales.

```python
pcs = pca.components_
pcs = pd.DataFrame(pcs)
features = df_PCA.columns
pcs.columns = features
pcs.index = [f"F{i}" for i in x_list]
pcs.round(2)
pcs.T
```

#### Nous pouvons désormais passer à la partie graphique avec une Heatmap permettant de visualiser les corrélations entre les paramètres d'origines et les composantes principales générées par l'ACP

```python
fig, ax = plt.subplots(figsize=(20, 6))
sns.heatmap(pcs.T, vmin=-1, vmax=1, annot=True, cmap="coolwarm", fmt="0.2f")
```

#### Nous n'étudirons ici que les composantes principales qui sont dé-corrélées entres elles et qui peuvent expliquer 80% de l'inertie de la variance soit les quatres premières composantes.
#### La première composante concentre une variabilité importante dans les données associées aux variable energy_100g, fat_100g et satured_fat_100g. Elle pourrait être simplifié en "Indice de Densité Énergétique et Lipidique".
#### La seconde composante concentre une variabilité importante dans les données associées aux variable salt_100g et sodium_100g. Elle pourrait être simplifié en "Indice de Densité Saline"
#### La troisième composante concentre une variabilité importante dans les données associées aux variable sugars_100g et carbohydrates_100g. Elle pourrait être simplifié en "Indice de Densité Sucré"
#### Et enfin la quatrième composante concentre une variabilité importante dans les données associées à la variable fiber_100g. Elle pourrait être simplifié en "Indice de Densité des Fibres"

#### Et enfin le cercle de corrélation permettant de mieux comprendre la structure des variables dans un jeu de données multidimensionnel. Nous pouvons identifier grâce à lui les paramètres qui influencent le plus les différentes composantes et effectuer des regroupements de paramètres suivant leur corrélation.
#### Ici nous nous intéressons uniquement aux deux premières composantes

#### Nous ne constatons pas de normalité pour cette variable, il faut alors réaliser une ANOVA non paramétrique.

#### Nous utilisons pour cela le test Kruskal-Wallis qui nous permet de déterminer si il existe des différences de distribution pour la variable sugars_100g.

```python
from scipy.stats import kruskal

groups = [group['sugars_100g'] for name, group in df.groupby('nutrition_grade_fr')]
h_statistic, p_value = kruskal(*groups)
print(f"Statistique H : {h_statistic}")
print(f"p-valeur : {p_value}")

alpha = 0.05  # Niveau de signification
if p_value < alpha:
    print("Les distributions des groupes sont différentes.")
else:
    print("Nous ne pouvons pas conclure que les distributions des groupes sont différentes.")
```
