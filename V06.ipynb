#### Nutri-Score = (Points positifs - Points négatifs) / Points positifs

##### Points positifs :

- Points pour les fruits, légumes, légumineuses x2
- Points pour les fibres alimentaires x2
- Points pour les protéines x2

##### Points négatifs :

- Points pour l'énergie (calories) /335
- Points pour les acides gras saturés /1.5
- Points pour les sucres simples ajoutés /4.5
- Points pour le sodium (sel) /0.09

```python
import pandas as pd
import numpy as np
from numpy import nan

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

#tools for testing our model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
X = df_without_Nan.iloc[:, 2:11]#toute la ligne de la 1ere à la 7eme colonne
y = df_without_Nan.iloc[:, 1]#c'est la réponse et donc on veut que la dernière colonne
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
#on garde 20% pour le test plus tard
```

```python
#mise à l'échelle des données
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
errors = []
for k in range(2,15):
    knn = KNeighborsClassifier(k)
    errors.append(100*(1 - knn.fit(X_train, y_train).score(X_test, y_test)))
plt.plot(range(2,15), errors, 'o-')
plt.show()
```

```python
#Define the model: Init K-NN
classifier = KNeighborsClassifier(n_neighbors=3, p=2, metric='euclidean')
```

```python
#fit model
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print(cm)
#print(f1_score(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
